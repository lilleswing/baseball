```python
%matplotlib inline  

import pandas as pd
import sqlite3

conn = sqlite3.connect('../baseball/baseball.db')
c = conn.cursor()
```

```python
q = "SELECT COUNT(event.id), game.year FROM event INNER join game on event.game_id=game.id where event.event like '%Batter Interference%' GROUP BY game.year"
interferences = pd.read_sql_query(q, conn)
```

```python
q = "SELECT COUNT(id), year from game group by year"
games = pd.read_sql_query(q, conn)
```

```python
interferences
```

```python
df = pd.merge(games, interferences)
```

```python
df.columns = ['Games', 'year', 'Interference Calls']
df = df[['year', 'Interference Calls', 'Games']]
```

```python
df['Interference Rate'] = df['Interference Calls'] / df['Games']
```
