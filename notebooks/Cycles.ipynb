```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('../baseball.db')
c = conn.cursor()
```

```python
def query(s):
    return [x for x in c.execute(s)]
```

```python
def csv_str(l):
    return ",".join([str(x) for x in l])
```

```python
cycle_events = ['Single', 'Double', 'Triple', 'Home Run']
results = []
for cycle_event in cycle_events:
    event_result = query("SELECT DISTINCT batter , game_id FROM event WHERE event='%s'" % cycle_event)
    results.append(event_result)
```

```python
counts = {}

for i in range(len(cycle_events)):
    for result in results[i]:
        if result not in counts:
            counts[result] = 0
        counts[result] += 1
cycles = []
for k,v in counts.items():
    if v == 4:
        cycles.append(k)
```

```python
cycles_df = pd.DataFrame(cycles)
cycles_df.columns = ['mlb_id', 'game_id']
```

```python
q = csv_str([x[1] for x in cycles])
games = pd.read_sql_query("SELECT * FROM game where id in (%s)" % q, conn)
q = csv_str([x[0] for x in cycles])
players = pd.read_sql_query("SELECT * FROM batter where mlb_id in (%s)" % q, conn)
```

```python
full_cycles = players.merge(cycles_df).merge(games, left_on=['game_id'], right_on=['id'])
full_cycles = full_cycles.drop(['id_x', 'mlb_id', 'id_y', 'game_id'], axis=1)
```
