```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('../baseball.db')
c = conn.cursor()
```

```python
df = pd.read_sql_query("SELECT * FROM event where event='Walk' and description like '%scores.%' \
                       AND event.pitcher in (SELECT mlb_id from pitcher)", conn)
```

```python
games = pd.read_sql_query("SELECT * FROM game", conn)
```

```python
df = df.merge(games, left_on='game_id', right_on='id')
```

```python
pitchers = pd.read_sql_query("SELECT * FROM pitcher", conn)
```

```python
df = df.merge(pitchers, left_on='pitcher', right_on='mlb_id')
```

```python
df = df[['id_x','name', 'timestamp']]
```

```python
counts = df.groupby('name').count().drop(['id_x'], axis=1)
counts.columns = ['counts']
```

```python
counts.sort_values(by='counts', ascending=False)
```
